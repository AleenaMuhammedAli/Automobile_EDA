### 1.Identify the 5 city has the highest number of sales


```python
df1=df.groupby(['CITY'],as_index=False).agg(Count=('CITY','count')).sort_values(by="Count",ascending=False).head(5)
df1 
```

```python
plt.figure(figsize=(8,9))
plt.pie(df1['Count'], labels=df1['CITY'], autopct='%1.1f%%', startangle=180)
plt.title('Top 5 Cities with the Highest Number of Orders')
plt.axis('equal') 
plt.show()
```

## Madrid has the highest count at 304, showing significant prominence among the listed cities. San Rafael and NYC follow, indicating regional differences in distribution, with Singapore and Paris having comparatively lower counts.

### 2.which country has the highest numbers of orders

```python
df2=df.groupby(['COUNTRY'],as_index=False).agg(Count=('COUNTRY','count')).sort_values(by="Count",ascending=False).head(5)
df2 
```

```python
sns.barplot(x='COUNTRY',y='Count',data=df2,palette='husl')
plt.title('Top 5 Country with the Highest Number of Orders')
```

## The USA leads with a count of 928, highlighting a substantial presence compared to other countries. Spain and France follow, while Australia and the UK show lower counts in the distribution.

### 3.which year has the highest numbers of orders

```python
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])
df['Year'] = df['ORDERDATE'].dt.year
year_count = df.groupby('Year').size().reset_index(name='Count')
year_count_sorted = year_count.sort_values(by='Count', ascending=False)
print(year_count_sorted)
```

```python
plt.figure(figsize=(5, 6))  
sns.barplot(data=year_count_sorted, x='Year', y='Count', palette='Blues_d')
plt.title('Number of Orders by Year', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Orders', fontsize=12)
plt.xticks(rotation=45) 
plt.tight_layout()
plt.show()
```

## The count peaked in 2019 at 1,329, showing notable growth from 2018's 953. However, there was a significant drop in 2020, with the count decreasing to 465.Because of the global pandemic covid-19 

### 4 List the top 10 months that generated the most revenue.

```python
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])
df['Year-Month'] = df['ORDERDATE'].dt.to_period('M')  
df4 = df.groupby('Year-Month')['SALES'].sum().reset_index().head(10)
df4['Year'] = df4['Year-Month'].dt.year  
df4['Month'] = df4['Year-Month'].dt.strftime('%B') 
df4 = df4[['Year', 'Month', 'SALES']]
df4_sorted = df4.sort_values(by='SALES', ascending=True)
print(df4_sorted)
```

```python
plt.figure(figsize=(10, 6))  
sns.barplot(data=df4_sorted, x='Month', y='SALES', hue='Year', palette='viridis')
plt.title('Top 10 Months generated the most revenue', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)
plt.xticks(rotation=45)  
plt.tight_layout()
plt.show()
```

## Sales in 2018 showed consistent growth, peaking dramatically in October at 448,452.95. This trend highlights a steady increase across most months, with significant jumps in the latter half of the year.

### 7.Which countries have purchased the highest number of products across different product lines, and what are the most popular product lines in those countries?


```python
country_product_count = df.groupby(['COUNTRY', 'PRODUCTLINE']).size() \
                          .unstack(fill_value=0) \
                          .assign(Total=lambda x: x.sum(axis=1)) \
                          .sort_values(by='Total', ascending=False) \
                          .drop(columns='Total') \
                          .head(5)

print(country_product_count.to_string())

country_product_count.plot(kind='bar', stacked=True, figsize=(10, 6))

plt.title('List of Products Purchased by Each Country')
plt.ylabel('Number of Products Purchased')
plt.xlabel('Country')
plt.xticks(rotation=45)
plt.show()
```

## The USA dominates the product line counts across all categories, particularly in Classic Cars with 311 units sold. Spain and France follow, showing varied preferences, while Australia and the UK have lower sales figures, especially in categories like Ships and Trains.

### 8. Identify the top 5 customers 

```python
top10customer = df['CUSTOMERNAME'].value_counts()[:5]
top10customer
```

```python

top10customer = df['CUSTOMERNAME'].value_counts()[:5]
top10customer_df = top10customer.reset_index()
top10customer_df.columns = ['CUSTOMERNAME', 'Order Count']
plt.figure(figsize=(10, 6))  
sns.barplot(data=top10customer_df, x='Order Count', y='CUSTOMERNAME', palette='viridis')
plt.title('Top 10 Customers by Order Count')
plt.xlabel('Number of Orders')
plt.ylabel('Customer Name')
plt.show()
```

### 9.Which country generates the most revenue?


```python
revenue_by_country = df.groupby('COUNTRY')['SALES'].sum().reset_index()
top_10_revenue_by_country = revenue_by_country.sort_values(by='SALES', ascending=False).head(5)
print(top_10_revenue_by_country)

plt.figure(figsize=(8, 7))  
sns.barplot(data=top_10_revenue_by_country, x='COUNTRY', y='SALES', palette='husl')
plt.title('Top 10 Countries by Sales Revenue', fontsize=14)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## The USA generates the highest sales at 3,355,575.69, indicating a strong market presence. In contrast, Spain and France follow with notable but significantly lower sales figures, while Australia and the UK have the least sales among the listed countries.

### 10.Which city generates the most revenue?


```python
revenue_by_CITY = df.groupby('CITY')['SALES'].sum().reset_index()
top_10_revenue_by_CITY = revenue_by_CITY.sort_values(by='SALES', ascending=False).head(5)
print(top_10_revenue_by_CITY)

plt.figure(figsize=(10, 6))  
sns.barplot(data=top_10_revenue_by_CITY, x='CITY', y='SALES', palette='husl')
plt.title('Top 10 CITY by Sales Revenue', fontsize=14)
plt.xlabel('CITY', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Madrid leads in sales with 1,082,551.44, highlighting its prominence among the cities listed. San Rafael and NYC also perform well, but Singapore and Paris show much lower sales figures, indicating regional disparities in market performance.

### 11.What is the median sales price for each product?

```python
df12 = df.groupby('PRODUCTLINE')['SALES'].median()
df12
```

```python
df12 = df12.reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(data=df12, x='PRODUCTLINE', y='SALES', palette='icefire')
plt.title('Median Sales Amount by Product Line')
plt.xlabel('Product Line')
plt.ylabel('Median Sales Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Classic Cars lead in average sales at 3,754.05, reflecting strong demand in that category. Meanwhile, Trucks and Buses and Motorcycles also show significant average sales, while Trains have the lowest average sales at 2,445.60, indicating varying levels of popularity across the product lines.

### 12.What is the mean sales price for each product?

```python
df13= df.groupby('PRODUCTLINE')['SALES'].mean()
df13
```

```python
df13 = df13.reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(data=df13, x='PRODUCTLINE', y='SALES', palette='pastel')
plt.title('Mean Sales Amount by Product Line')
plt.xlabel('Product Line')
plt.ylabel('Mean Sales Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Classic Cars have the highest sales figure at 4,049.39, indicating robust performance in that category. In contrast, Trains show the lowest sales figure at 2,938.23, suggesting varying levels of demand across the different product lines.
